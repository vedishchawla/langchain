### model integration like groq gemini gpt

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
os.environ["GEMINI_API_KEY"]=os.getenv("GEMINI_API_KEY")
```

```python
from langchain_core.messages.block_translators import google_genai
from langchain_core.prompts import message
from langchain.chat_models import init_chat_model

model = init_chat_model(model="google_genai:gemini-2.5-flash-lite")
model
```

```python
response = model.invoke("Hello how are u give a short brief abotu java script")
response
```

```python
print(response.content)
```
