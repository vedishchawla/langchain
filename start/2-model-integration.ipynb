# Streaming

dont wait for complete response and then display

while getting the response start showing it

## better usser experince

```python
for chunk in model.stream("Tell me about javascript in 1000 words"):
    print(chunk.text,end="",flush=True)
```

# Batch

### batches me question bhejo model ko reduces cost 

```python
reponses = model.batch([
    "tell about js in short",
    "tell about python in short",
    "tell about competitive programming in short"
])
for response in reponses:
    print(response.content)
    print("---------------*****--------------------------------*****--------------------------------*****-----------------")
```
